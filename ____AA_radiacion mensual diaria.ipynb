```python
def boxplots_meses(tmy):   
    # tmy = df.copy()
    tmy['dti'] = pd.date_range("2020-01-01", periods=365 * 24, freq="H")
    tmy = tmy.set_index('dti')
    tmy['D'] = tmy.index.date
    tmy['H'] = tmy.index.hour

    df = tmy.pivot(index=['D'], columns='H', values='G(h)')
    df.index = pd.to_datetime(df.index)
    fig,[ax12,ax3,ax7]=plt.subplots(nrows=1, ncols=3,figsize=(16,3))

    df[(df.index.month==12)].boxplot(ax=ax12);
    df[(df.index.month==3)].boxplot(ax=ax3);
    df[(df.index.month==7)].boxplot(ax=ax7);
    ax12.set_ylim(0,1000);ax3.set_ylim(0,1000);ax7.set_ylim(0,1000);

    plt.close()
    fig = mpl_to_html(fig)
    return fig

    
fig=boxplots_meses(tmy)    
display(HTML(Markup(f'<center><img src={fig}></center>')))
```
